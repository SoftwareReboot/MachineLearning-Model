```python
X = df[['Age', 'Annual Salary', 'Credit Card Debt', 'Net Worth']]  # features
y = df['Car Purchase Amount']  # target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=10)
```

```python
print("Training set", X_train.shape)
print("Testing set", X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
numeric_features = [ 'Age', 'Annual Salary', 'Credit Card Debt', 'Net Worth']
X_train[numeric_features] = scaler.fit_transform(X_train[numeric_features])
X_test[numeric_features] = scaler.transform(X_test[numeric_features])
```

```python
#Build the Classifier: import the Linear Regression library
from sklearn.linear_model import LinearRegression
```

```python
#create A Linear Regression model
lm = LinearRegression()
```

```python
#Train the data model
lm.fit(X_train,y_train)
```

```python
#get the intercept
intercept = lm.intercept_

#display the Intercept
print (f"Intercept : {intercept: .2f}")
```

```python
#get the slope of each independent variable (features)

coeff_df= pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient'])
```

```python
#show the coefficient
print (coeff_df)
```

Predict the total amount that customers are willing to pay for the new car using the model.

```python
#Predict Car Purchase Amount using the training data
y_pred_train = lm.predict(X_train)
y_pred_train
```

```python
# Predict the car purchase amount using your own data
# Example values:
# Age = 35
# Annual Salary = 75000
# Credit Card Debt = 5000
# Net Worth = 60000

lm.predict([[35, 75000, 5000, 60000]])[0]
```

```python
# Predict on the test set
y_pred_test = lm.predict(X_test)

# Now create the comparison DataFrame
y_pred_df = pd.DataFrame({
    'Actual Car Purchase Amount': y_test,
    'Predicted Car Purchase Amount': y_pred_test,
    'Difference': y_test - y_pred_test
})

# Show first 20 rows
y_pred_df[0:20]
```

```python
#visualize the difference between actual and prediction results.

import matplotlib.pyplot as plt

#plt.figure(figsize=(10,10))
plt.scatter(y_train, y_pred_train, label= 'Training data',color='b', alpha=0.7)
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], 'r', label='Regression line')


plt.legend()
plt.xlabel('Actual Car Purchase Amount')
plt.ylabel('Predicted Car Purchase Amount')
plt.title('Actual Price vs Predicted Price')
plt.show()
```

Evaluate the model

```python
#r2 measures how close are the data fitted to the regression line
# r_square answers the question: Is the model is good or bad?

from sklearn.metrics import r2_score

score = r2_score(y_train,y_pred_train) # portion of the variance that can be explained by the training features
print (f"Score: {score*100: .2f}%") #coeff of determination
```
